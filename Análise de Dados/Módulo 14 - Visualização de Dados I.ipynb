## 1\. Preço do diamante por tipo de corte

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = data[['cut','price']].groupby('cut').agg('mean').reset_index()
print(diamonds)

grafico_escolhido = sns.barplot(data=diamonds, x='cut', y='price', errorbar=None, palette='dark')
grafico_escolhido.set(title='Preço do diamante por tipo de corte', xlabel ='Corte', ylabel = 'Preço médio (USD)')
```

**Insight do gráfico 1**: ...



*   O corte PREMIUM tem o maior preço.
*   O corte IDEAL tem o menor preço.
*   Os cortes Good e Very Good tem preços semelhantes.

## 2\. Número de passageiros em dezembro por ano

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 2

flights = data.loc[data['month']=='Dec', ['year', 'passengers']].reset_index()
print(flights)

grafico_escolhido = sns.lineplot(data=flights, x='year', y='passengers', marker='o')
grafico_escolhido.set(title='Número de passageiros em dezembro por ano', xlabel ='Ano', ylabel = 'Número de passageiros')
```

**Insight do gráfico 2**: ...



*   Conforme a progressão dos anos, em dezembro,  número de passageiro veio crescendo.

*   Entre 1957 e 1958 houve um crescimento mínimo no número de passageiros.

## 3\. Numero de passageiros por mês entre 1949 e 1959

```
import seaborn as sns

data = sns.load_dataset("flights")
data.head()
```

```
# gráfico do exercício 3

flights = data.query("1949 <= year < 1960")

with sns.axes_style('whitegrid'):
  grafico_escolhido = sns.lineplot(data=flights, x='month', y='passengers', hue='year', palette='dark')
  grafico_escolhido.set(title='Número de passageiros por mês entre 1949 e 1960', xlabel = 'Mês', ylabel = 'Passageiros')
  grafico_escolhido.get_legend().set_title('Ano')
```

**Insight do gráfico 3**: ...



*   Os meses de julho e agosto consistentemente apresentam os picos mais altos de passageiros.

*   Os meses de janeiro e fevereiro geralmente registram o menor número de passageiros em cada ano.

*   Há uma tendência geral de aumento no número de passageiros ao longo dos anos.
