Nestes exercícios, você deve decidir qual é o gráfico visto em aula que melhor visualiza uma base de dados. Após decidir, você deverá criar a visualização usando o conteúdo exposto durante a aula e adicionar um pequeno parágrafo sobre um insights que pode ser extraido do gráfico.

## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

with sns.axes_style('darkgrid'):
  diamonds = sns.boxplot(x=data['price'])
  diamonds.set(title='Preços outliers de diamante', xlabel='Preço (USD)')
```

**Insight do gráfico 1**: ...



*   A maioria dos diamantes tem preço abaixo de 5.000 dólares.
*   Temos aqui uma distribuição assimétrica, com muitos outliers.

## 2\. Correlação entre o do preço do diamante com seu peso (`carat`) agrupados por sua transparêcia (`clarity`)

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 2

with sns.axes_style('darkgrid'):
  diamonds = sns.scatterplot(data=data, x="carat", y="price", hue="clarity", alpha=0.6)
  diamonds.set(title="Preço vs Peso por Claridade", xlabel="Peso (carat)", ylabel="Preço (USD)")
  diamonds.get_legend().set_title("Transparência")
```

**Insight do gráfico 2**: ...



*   Existe uma correlação entre peso e preço do diamante. Mostrando que diamantes com menor clareza tendem a se concetrar em preços menores.

*   O peso dos diamantes mantiveram bastante semelhantes, concentrando-se em sua maioria abaixo de 03.

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

cars = sns.load_dataset("mpg")

with sns.axes_style('darkgrid'):
  cars = sns.histplot(data=cars, x="weight", fill=True, color="blue", kde=True)
  cars.set(title="Distribuição do Peso dos Carros", xlabel="Peso (lbs)", ylabel="Contagem")
```

**Insight do gráfico 3**: ...



*   A maior densidade de carros está entre 2.000 e 4.000 libras.
*   Há uma indicação de que a maior parte dos modelos analisados pertence a uma faixa de peso intermediária.
*  O peso com a menor densidade é o de 5.000 libras.
