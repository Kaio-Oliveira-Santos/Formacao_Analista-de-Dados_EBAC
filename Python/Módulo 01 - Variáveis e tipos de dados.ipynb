Extraia da string abaixo o valor da taxa **selic** na variável `selic` e o valor do **ano** na variavel `ano`. Imprima os valores na tela.

```
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'
```

```
comeco_da_taxa = noticia.find('Selic vai a ') + len('Selic vai a ')
final_da_taxa = noticia.find('%', comeco_da_taxa) + 1
taxa_selic = noticia[comeco_da_taxa:final_da_taxa]

inicio_ano = noticia.find('alta em ') + len('alta em ')
fim_ano = noticia.find(' anos', inicio_ano) + len(' anos')
ano = noticia[inicio_ano:fim_ano]

print("Taxa Selic: ", taxa_selic)
print("Ano: ", ano)
```
