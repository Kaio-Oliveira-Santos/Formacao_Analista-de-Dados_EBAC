**Desafio: **

Faça um programa que calcule o **valor total investido por um colecionador em sua coleção de CDs** e o **valor médio gasto em cada um deles**. O usuário deverá informar a quantidade de CDs e o valor para cada um.

Lembre-se de que você já tem a habilidade de desenvolver variáveis, estruturar dados, criar condições, repetições e funções.

```python
# Criando a função:
def valor_total_investido():

# Crio uma varíavel para armazenar o número de CD's da coleção:
  quantidade_cds = int(input("Informe a quantidade de CDs: "))

# Crio uma varíavel para armazenar o total investido, começando em R$ 0.
  total_investido = 0

# Crio um laço for para perguntar o valor de cada CD baseado no número de CD's informados anteriormente.
  for CD in range(quantidade_cds):
    valor_cd = float(input(f'Informe o valor do CD {CD + 1}: '))

# Adiciono ao total investido a soma de cada valor informado.
    total_investido += valor_cd

# Crio uma condicional IF/ELSE para calcular o valor médio, se a quantidade de CD's for maior que zero.
  if quantidade_cds > 0:
    valor_medio = total_investido / quantidade_cds
  else:
    valor_medio = 0

  print(f'\nTotal investido na coleção: R$ {total_investido:.2f}')
  print(f'\nValor média da coleção: R$ {valor_medio:.2f}')

# Chamando a função:
valor_total_investido()
```
